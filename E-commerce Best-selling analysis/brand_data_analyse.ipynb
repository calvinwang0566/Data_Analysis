```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv("brand_data.csv")
df = df[df["avg_reviews"] > df["avg_reviews"].mean()]

print(df["avg_price"].describe())

df = df.sort_values(by = "avg_rating").reset_index(drop = True)
```

```python
x = np.array(df["avg_price"])
y = np.array(df["total_products"])
z = np.array(df["avg_rating"])

fig = plt.figure()
fig.add_subplot(111, projection = "3d")
ax = Axes3D(fig)
ax.scatter(x, y, z)
ax.set_xlabel("price")
ax.set_ylabel("product num")
ax.set_zlabel("rating")
```
